- KNN 모델 (K-Nearest Neighbors) : 유유 상종의 개념과 비슷
    - 새로운 데이터가 들어올때 거리계산을 통해 가장 가까운 n개의 특성의 데이터로 에측 하는 모델
- KNN 모델의 이웃의 숫자를 조정(하이퍼파라미터 튜닝) -> 일반화가 이루어지는 방향으로조정

### 1. 문제 정의 (목표)
- 붓꽃의 4개의 특성을 통해 붓꽃의 품종을 예측하는 모델
- 특성 4개 : 꽃잎의 길이와 너비, 꽃받침의 길이와 너비
- 정답 : 클래스 3개 -붓꽃의 3개지 품종 (setosa, versicolor, virginica)

```python
# 라이브러리 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# 학습용 iris 데이터셋 로드
from sklearn.datasets import load_iris
```

### 2. 데이터 수집
- 학습용 으로 제작된 iris데이터로드

```python
# 번치 객체 딕셔너리와 비슷한 자료구조
# 묶음, 다발 이라는 의미 >> 머신러닝에서 주로 사용되는 데이터 셋

# data는 문제데이터
# target : 정답 데이터의 
# target_name : 정답 데이터 이름
# feature_names : 특성의 이름
# DESCR : 데이터의 설명 
iris_data = load_iris()
iris_data
```

- 우리에게 익숙한 데이터 프레임 형태로 문제 데이터 변환

```python
iris_df = pd.DataFrame(iris_data['data'], columns = iris_data['feature_names'])
iris_df
```

```python
# 문제와 정답을 선언
X = iris_df # 문제 데이터
y = iris_data['target'] # 정답 데이터
```

- 학습 평가 데이터로 최종 분리
    - 랜덤하게 뽑아야 한다

```python
# 학습/테스트 데이터 분리를 위한 임포트
from sklearn.model_selection import train_test_split
```

```python
# 알아서 훈련용과 테스트용 데이터를 나누어 주는 도구 (랜덤)
# 순서 중요
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    test_size=0.3,  # 30%가 테스트 데이터
    random_state=8 # 랜덤 시드 고정
)
```

#####  train_test_split
1. train test분리
2. 랜덤 샘플링이 계속 되버리면 학습 데이터가 변경 되므로 하이퍼파라미터 조정의 의미가 없어짐
3. 랜덤 시드를 고정 (수행시마다 동일한 샘플링 결과를 얻기 위해)
4. 랜덤 시드는 큰 의미 없고 내부적 규칙을 가지는 숫자

#### 산점도 매트릭스로 시각화 해보기
- x축과 y축에 특성을 넣고 출력하는 산점도 여려개를 한번에 출력
- 여러 특성쌍의 관계를 동시에 확인할 수 있음

```python
pd.plotting.scatter_matrix(iris_df,
                          figsize=(15,15), # 매트릭스 사이즈 조절
                          marker='o', # 데이터의 표현 형태  동그라미 s 네모
                          c = y, # 데이터의 색상 
                          alpha =  1  # 매트리스 투명도 설정 1은 불투명
                          )
```

#### 5. 모델링(모델 선택 학습 평가)

```python
# 모델 정의
knn_model = KNeighborsClassifier(n_neighbors=18)
```

```python
# 모델 학습
knn_model.fit(X_train, y_train)
# pred = knn_model.predict(X_test)
# pred
```

```python
# 모델 평가
knn_model.score(X_train, y_train) # k = 18 일떄 0.9619047619047619
```

```python
knn_model.score(X_test, y_test) # k = 18 일때 1
```

- 반복문 사용하여 하이퍼파라미터 튜닝하기

```python
train_list = [] # train 데이터의 정확도
test_list = [] # test 데이터의 정확도
# n_neighobrs 1 ~ 49
# for, append
for i in range(1, 50, 1):
    # 모델 정의
    knn_model = KNeighborsClassifier(n_neighbors = i) # 하이퍼파라미터 k값을 1 ~ 49까지 넣어봄

    # 모델 학습
    knn_model.fit(X_train, y_train)

    # 훈련 데이터의 예측 정확도를 리스트에 담기
    train_score = knn_model.score(X_train, y_train)
    train_list.append(train_score)
    
     # 평가 데이터의 예측 정확도를 리스트에 담기
    test_score = knn_model.score(X_test, y_test)
    test_list.append(test_score)
    
```

```python
train_list
```

```python
test_list
```

```python
# 그래프를 그려서 확인해보자
# 각각의 하이퍼 파라미터 따른 훈련데이터의 정확도, 테스트 데이터의 정확도를 그래프를 그려서
# 일반화가 되고있는 최적점(하이퍼 파라미터 값)이 어디인지확인해보자

plt.figure(figsize=(13,5))

# range(1, 50, 1) x 축에 사용할 값
# train_list y 축에 사용할 값들 (k 값에 따른 정확도)
# label = 'tarain_acc': 범려에 표시될 텍스트
plt.plot(range(1, 50, 1), train_list, label = 'tarain_acc')

plt.plot(range(1, 50, 1), test_list, label = 'test_acc')

plt.legend() # 범례도 표시하고
plt.grid() # 눈금선

plt.xticks(range(1, 50)) # 눈금선 간격

plt.show()
```

```python
# 모델 복잡도 곡선을 확인 해본 결과, 하이퍼파라미터가 7,8,10 가장 높다
# test 데이터가 예측 정확도가 가장 높은걸 확인 할수있다
# k=8 , k=10 인경우 
# test 정확도 > train 정확도
# 신뢰성이 떨어진다.
# k=7 일 경우 test 정확도, train 정확도 둘다 높구 차이가 미비하다

# 모델 생성 할 때 현재 우리가 가지고 있는 랜던 샘플링된 데이터의 적합한 하이퍼파라미터 는 7 이라는걸 알수 있다
# 단, 이 데이터들은 랜덤 시드 값에 따라 달라 질수 있음
```
