#### 성격데이터 생성

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(
    {'시간' : [2,4,8,9],
    '성적' : [20,40,80,90]}
)
data
```

w와 b를 찾는 방법 
- 최소 제곱법 : 수학적 공식을 이용한 해석적 방법
- 경사 하강법 : w와 b값의 변화의 따른 오차값의 (MSE) 기울기를 이용하여 찾는 방법

 최소제곱법이 적용된 sklearn 모델을 사용해보기

```python
from sklearn.linear_model import LinearRegression
```

```python
# 모델 생성
lr_rg = LinearRegression() # 하이퍼 파라미터 없음
```

```python
# 학습
# 문제 독립
# 정답 종속
lr_rg.fit(data[['시간']],data['성적'])
```

```python
# 예측 
lr_rg.predict([[7]])
```

```python
# 가중치 (w) 절편 (b) 확인
lr_rg.coef_
```

```python
lr_rg.intercept_
```

#### 경사하강법(Gradient Descent)
1. 파라미터 (w, b) 초기화 : 임의의 값으로
2. 예측 : 현재 w와 b를 사용하여 예측을 수행
3. 오차 계산 : 예측값과 실제 값 사이의 오차 계산
4. 비용함수의 기울기 계산 : 오차값이 낮아지는 방향을 찾기 위해서, w와 mse간의 기울기를 값을 계산
5. 파라미터 업데이트 : 기울기와 학습률을 사용해서 업데이트

```python
# 가설 함수 예측값을사용
def h (w,x):
    return w*x+0
```

```python
# MSE(평균제곱오차) : 손실함수, 비용함수
def cost(data,target,weight): # 특성(시간), 실제 값(실제 성적), 가중치
    y_pred = h(weight,data)
    mse = ((target - y_pred)**2).mean()
    return mse
```

```python
# MSE
cost(data['시간'],data['성적'], 10)
```

- w값에 따른mse 변화율 그려보기

```python
# x축의 범위
w_arr = range(-10,31)
```

```python
cost_list = [] # 비용 함수 MSE 담아줄 리스트
for w in w_arr:
    c = cost(data['시간'],data['성적'], w)
    cost_list.append(c)
```

```python
cost_list
```

```python
plt.plot(w_arr, cost_list)
plt.xlabel('weight')
plt.ylabel('loss')
plt.show()
```

```python
# 경사 하강법이 적용 된 모델
from sklearn.linear_model import SGDRegressor
```

```python
# 모델 초기화( 생성 )
sgd_rg = SGDRegressor(
    eta0 = 0.01, # 학습률 기울기의 보폭
    max_iter = 5000, # w값을 찾아가는 업데이트 하는 횟수를 최대 5000번까지 제한
    verbose = 1 # 학습 진행 현황 출력
)
```

```python
# 학습
sgd_rg.fit(data[['시간']], data['성적'])
```

```python
# 가중치 ,절편확인
sgd_rg.coef_, sgd_rg.intercept_
```

```python
# 성능 확인
sgd_rg.score(data[['시간']], data['성적'])
# 회귀 평가 지표 (오차값, 데이터의 분포를 잘 이해하고 있는가)
```
