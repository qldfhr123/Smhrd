#### 선형회귀 모델 데이터 특성 확장
- 선형회귀 모델은 데이터의 분포가 직선의 형태를 많이 벗어나게 되면서 예측 정확도가 낮아짐
- 선형회구 모델은 모델 튜닝을 위한 하이파파라미터가 존재하지않음
- 특성확장을 통해 성능을 높여보자

```python
# copy는 기존의 데이터는 그대로 두면서, 복사를 해서 새로운 변수명에 넣어줌
# copy 없이 X_train을 새로운 변수에 담아준다면 변수의 내용이 바뀔 경우 기존 X_train도 바뀔수있다
# 주소값 관련
extended_X_train = X_train.copy()
```

```python
# 13 번 반복
for col1 in X_train.columns:
    for col2 in X_train.columns:
        extended_X_train[col1+'x'+col2] = X_train[col1] * X_train[col2] # 169번 새로운 열 추가
```

```python
extended_X_train
```

```python
extended_X_test = X_test.copy()
```

```python
# 13 번 반복
for col1 in X_test.columns:
    for col2 in X_test.columns:
        extended_X_test[col1+'x'+col2] = X_test[col1] * X_test[col2] # 169번 새로운 열 추가
```

```python
extended_X_test
```

```python
# 특성 확장된 train 데이터로 학습
linear_model2 = LinearRegression()
linear_model2.fit(extended_X_train, y_train)
```

```python
linear_model2.score(extended_X_train, y_train) # 특성 확장전 : 0.76...
```

```python
linear_model2.score(extended_X_test, y_test) # 특성 확장전 : 0.67...
```

라쏘(Lasso) 모델 적용

```python
from sklearn.linear_model import Lasso
```

```python
# 알파 값이 높아지면 규제가 강해지면서 과대 적합은 줄더듬 오차는 커질수있음
lasso_model = Lasso() # 알파 값 default 값 : 1
```

```python
lasso_model.fit(extended_X_train, y_train)

```

```python
sum(lasso_model.coef_ != 0)
# 182개의 특성중 94개 특성이 남았다
```

```python
lasso_model.score(extended_X_train, y_train) # 선형 회귀 0.93...
```

```python
lasso_model.score(extended_X_test, y_test) # 선형 회귀 0.81...
# 테스트에서 올랐다 >> 과대 적합 해소 (train, test 차이가 줄어들었다)
```

- train 에서 R2 스코어는 내려갔지만 test에서는 오름
- 선형회귀 모델에 비해 과대 적합이 줄어드는 것을 확인할수있다

릿지 모델 적용

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge() # 알파 값 default 값 : 1
```

```python
ridge_model.fit(extended_X_train, y_train)
```

```python
ridge_model.score(extended_X_train, y_train) # 라쏘 0.88..
```

```python
ridge_model.score(extended_X_test, y_test) # 라쏘 0.84..
```

- 알파값 규제값 1일때, 릿지 모델에 비해 라쏘 모델이 과대 적합 해소 에는 더 적합하다!

#### 하이퍼 파라미터 튜닝
- alpha 값을 바꿔가며 RMSE를 확인

```python
# lasso 모델
# alpha 값에 따라 RMSE 계산값을 빈 리스트에 담아주기
alpha_list = [0.001, 0.01, 0.1 , 1, 10, 100]
lasso_train_list = []
lasso_test_list = []

for i in alpha_list:
    # alpha 값을 변경 하면서 라쏘 모델 불러오기
    lasso_model = Lasso(alpha = i)
    lasso_model.fit(extended_X_train, y_train)

    # 예측값 구하기
    lasso_train_pre = lasso_model.predict(extended_X_train)
    # 예측값과 정답(실제값)을 이용해서 RMSE 추출
    lasso_train_RMSE = mean_squared_error(lasso_train_pre, y_train)**0.5
    # 빈 리스트에 RMSE 값 넣어주기
    lasso_train_list.append(lasso_train_RMSE)

    # test 
    lasso_test_pre = lasso_model.predict(extended_X_test)
    # 예측값과 정답(실제값)을 이용해서 RMSE 추출
    lasso_test_RMSE = mean_squared_error(lasso_test_pre, y_test)**0.5
    # 빈 리스트에 RMSE 값 넣어주기
    lasso_test_list.append(lasso_test_RMSE)
```

```python
lasso_train_list
```

```python
lasso_test_list 
```

```python
# 한글을 쓰기 위한 rc import
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
plt.figure(figsize = (10, 5))
plt.plot(lasso_train_list, label = 'train')
plt.plot(lasso_test_list, label = 'test')
plt.title('Lasso 모델 : alpha - RMSE 그래프')
plt.xlabel('alpha값')
plt.xticks(np.arange(6), alpha_list) # x 축의 tick 값을 설정 (x축 값 개수, 실제 들어갈 값)
plt.ylabel('RMSE')
plt.legend()
plt.grid()
plt.show()
```

```python
# ridge 모델
# alpha 값에 따라 RMSE 계산값을 빈 리스트에 담아주기
alpha_list = [0.001, 0.01, 0.1 , 1, 10, 100]
ridge_train_list = []
ridge_test_list = []

for i in alpha_list:
    # alpha 값을 변경 하면서 모델 불러오기
    ridge_model = Ridge(alpha = i)
    ridge_model.fit(extended_X_train, y_train)

    # 예측값 구하기
    ridge_train_pre = ridge_model.predict(extended_X_train)
    # 예측값과 정답(실제값)을 이용해서 RMSE 추출
    ridge_train_RMSE = mean_squared_error(ridge_train_pre, y_train)**0.5
    # 빈 리스트에 RMSE 값 넣어주기
    ridge_train_list.append(ridge_train_RMSE)

    # test 
    ridge_test_pre = ridge_model.predict(extended_X_test)
    # 예측값과 정답(실제값)을 이용해서 RMSE 추출
    ridge_test_RMSE = mean_squared_error(ridge_test_pre, y_test)**0.5
    # 빈 리스트에 RMSE 값 넣어주기
    ridge_test_list.append(ridge_test_RMSE)
```

```python
plt.figure(figsize = (10, 5))
plt.plot(ridge_train_list, label = 'train')
plt.plot(ridge_test_list, label = 'test')
plt.title('Ridge 모델 : alpha - RMSE 그래프')
plt.xlabel('alpha값')
plt.xticks(np.arange(6), alpha_list) # x 축의 tick 값을 설정 (x축 값 개수, 실제 들어갈 값)
plt.ylabel('RMSE')
plt.legend()
plt.grid()
plt.show()
```

- Lasso 모델과 Ridge 모델 모두 alpha 값이 10일때 가장 과대 적합 제어에 유리한것을 볼 수 있음
- Lasso 모델은 Ridge 모델에 비해, 과대 적합 제어가 더 잘된것을 확인
- 과대 적합 제어를 위해서는 Lasso을 사용하는 것을 유리하다
- 그러니 규제의 정도에 따라 Lasso모델은 오차값이 급격하게 증가함을 확인
- 모델의 정확한 예측 (r2 score)면에서는 Ridge모델이 유리하다
