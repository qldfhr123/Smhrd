```python
# 크롬열어서
driver = wb.Chrome()
driver.get("https://www.google.com/search?sca_esv=e5d69723eb1b552b&q=%EB%94%94%EC%A7%80%EB%AA%AC&udm=2&fbs=ABzOT_CZsxZeNKUEEOfuRMhc2yCI6hbTw9MNVwGCzBkHjFwaK53DgNHTMxn53_XGiUHS2MvCJOyRBFAs1hn9fBmZ19LYQ9yir0OaCtXuvoNzG8-dHms3wc7E7cm3bLVDfZkvpd-9FL0Ome2QZ2kvWuCs0U3qzTBxUhx0Kwtj3j_SllEPpRfq0AFFpotpqEUon0k1z37H_kXzjxTeYy1VisfJ_U-6GRUmLg&sa=X&ved=2ahUKEwj5_oDfsdmMAxV8sFYBHbacIVsQtKgLegQIFhAB&biw=1536&bih=695&dpr=1.25")
```

```python
# 스크롤
body = driver.find_element(By.CSS_SELECTOR,"body")

for i in range(10):
    body.send_keys(Keys.END)
    time.sleep(1)
```

```python
# 이미지 가져오기
img = driver.find_elements(By.CSS_SELECTOR, "a>div>div>div>g-img>img")
```

```python
len(img)
```

```python
img_src = []

# 요소에서 속성 가져오기
for i in range(len(img)):
    img_src.append(img[i].get_attribute("src"))
```

```python
img_src
```

```python
# src 값을 주면 실제 파일로 저장
from urllib.request import urlretrieve
```

```python
for i in range(50):
    urlretrieve(img_src[i],"C:/Users/smhrd/Desktop/이미지/"+str(i)+".jpg")
```
