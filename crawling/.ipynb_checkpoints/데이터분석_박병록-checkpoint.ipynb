```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
# Question2
driver = wb.Chrome()
driver.get("https://www.naver.com/")
search = driver.find_element(By.ID, "query")
search.send_keys("크롤링")
search.send_keys(Keys.ENTER)
```

```python
# Question3
# 전부 가져오기
driver = wb.Chrome()
driver.get("https://www.11st.co.kr/browsing/BestSeller.tmall?method=getBestSellerMain&xfrom=main^gnb#pageNum%%3")

body = driver.find_element(By.CSS_SELECTOR,"body")
pro = driver.find_elements(By.CSS_SELECTOR, ".viewtype.catal_ty > .cfix > li > div > a")

title_list = []
price_list = []
review_list = []


for i in range(0,6,1):
    pro = driver.find_elements(By.CSS_SELECTOR, ".viewtype.catal_ty > .cfix > li > div > a")
    pro[i].click()
    time.sleep(1)
    title = driver.find_element(By.CSS_SELECTOR, "h1.title")
    title_list.append(title.text)
    price = driver.find_element(By.CSS_SELECTOR, "#finalDscPrcArea > dd.price > strong > span.value")
    # 특가 할인 상품을 위한 조건문
    if not price.text.strip():
        price = driver.find_element(By.CSS_SELECTOR, "#maxDiscountResult > dd.price > strong > span.value")
    price_list.append(price.text)
    
    review = driver.find_element(By.CSS_SELECTOR, "#layBodyWrap > div > div.s_product.s_product_detail > div.l_product_cont_wrap > div > div.l_product_view_wrap > div.l_product_summary > div.l_product_side_info > div.b_product_info_assist > div").text
    if not review:
        review = "리뷰가 존재하지 않음"
        
    elif review != "리뷰가 존재하지 않음":
        review = driver.find_element(By.CSS_SELECTOR, "#prdReview > strong").text
    
    review_list.append(review)
    driver.back()
    time.sleep(1)


dic = {"상품" : title_list, "가격" : price_list, "리뷰수" : review_list}
store = pd.DataFrame(dic)
```

```python
# Question4
# 전부 가져오기
driver = wb.Chrome()
driver.get("https://www.melon.com/chart/index.htm")

rank = driver.find_elements(By.CSS_SELECTOR, "#lst50 > td:nth-child(2) > div > span.rank, #lst100 > td:nth-child(2) > div > span.rank")
singer = driver.find_elements(By.CSS_SELECTOR, "div.ellipsis.rank02 > span")
song = driver.find_elements(By.CSS_SELECTOR, "div.ellipsis.rank01 > span > a")
like =driver.find_elements(By.CSS_SELECTOR, "#lst50 > td:nth-child(8) > div > button > span.cnt, #lst100 > td:nth-child(8) > div > button > span.cnt")

rank_list =[]
singer_list = []
song_list = []
like_list = []


for i in range(len(singer)):
    rank_list.append(rank[i].text)
    singer_list.append(singer[i].text)
    song_list.append(song[i].text)
    like_list.append(like[i].text)


dic = {"순위" : rank_list, "곡명" : singer_list, "가수" : song_list, "좋아요" : like_list}
melon = pd.DataFrame(dic)
melon.to_csv("멜론TOP100.csv",encoding="euc-kr")
```
