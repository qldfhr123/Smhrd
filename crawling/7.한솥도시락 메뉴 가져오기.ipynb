```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

import time
```

```python
driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")

btn = driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a") 
for i in range(5):
    btn.click()
    time.sleep(1)

# 리스트 형태
menu = driver.find_elements(By.CSS_SELECTOR, ".item-text > h4")
price = driver.find_elements(By.CSS_SELECTOR, ".item-text > .item-price")

menu_list = []
price_list = []

for i in range(len(menu)):
    menu_list.append(menu[i].text.strip())
    price_list.append(price[i].text.strip())


dic = {"도시락 메뉴" : menu_list, "가격" : price_list}
han = pd.DataFrame(dic)
# han.to_csv("한솥 메뉴.csv",encoding="euc-kr")
```
